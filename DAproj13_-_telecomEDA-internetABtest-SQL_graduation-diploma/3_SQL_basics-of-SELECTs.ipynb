## Описание базы данных

### Таблица books
Содержит данные о книгах:
* `book_id          —` идентификатор книги (`PK`);
* `author_id        —` идентификатор автора (`FK`);
* `title            —` название книги;
* `num_pages        —` количество страниц;
* `publication_date —` дата публикации книги;
* `publisher_id     —` идентификатор издателя (`FK`).

### Таблица authors
Содержит данные об авторах:
* `author_id        —` идентификатор автора (`PK`);
* `author           —` имя автора.

### Таблица publishers
Содержит данные об издательствах:
* `publisher_id     —` идентификатор издательства (`PK`);
* `publisher        —` название издательства;

### Таблица ratings
Содержит данные о пользовательских оценках книг:
* `rating_id        —` идентификатор оценки (`PK`);
* `book_id          —` идентификатор книги (`FK`);
* `username         —` имя пользователя, оставившего оценку;
* `rating           —` оценка книги.

### Таблица reviews
Содержит данные о пользовательских обзорах на книги:
* `review_id        —` идентификатор обзора (`PK`);
* `book_id          —` идентификатор книги (`FK`);
* `username         —` имя пользователя, написавшего обзор;
* `text             —` текст обзора.

# Выполнение экзаминационных вопросов

## Запрос №1: Посчитайте, сколько книг вышло после 1 января 2000 года

```python
query = '''
SELECT COUNT(book_id) AS books_issued_after_Jan1_2000
FROM books
WHERE publication_date > '2000-01-01'
;'''

# Комментарий к запросу: поскольку book_id - это primary key таблицы, то DISTINCT излишен

output_1 = pd.io.sql.read_sql(query, con = engine)
output_1
```

## Запрос №2: Для каждой книги посчитайте количество обзоров и среднюю оценку

```python
query = '''
WITH
brevno AS (SELECT book_id, COUNT(review_id)
           FROM reviews
           GROUP BY book_id),
bratno AS (SELECT book_id, AVG(rating)
           FROM ratings
           GROUP BY book_id)

SELECT DISTINCT books.book_id, 
       brevno.count AS nof_reviews,
       bratno.avg AS avg_rating
FROM books,brevno,bratno
WHERE books.book_id = brevno.book_id AND books.book_id = bratno.book_id
;'''

# Комментарий к запросу: не совсем как нас учили, зато так логика прозрачна на 100% 
# ведь SQL не язык программирования, а язык описания требований к данным!

output_2 = pd.io.sql.read_sql(query, con = engine)
output_2
```

## Запрос №3: Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры

```python
query = '''
SELECT publisher AS top_publisher
FROM publishers
WHERE publisher_id = (SELECT publisher_id
                      FROM (SELECT publisher_id, COUNT(book_id)
                            FROM books
                            WHERE num_pages > 50
                            GROUP BY publisher_id
                            ORDER BY count DESC
                            LIMIT 1 
                           ) AS dummy_alias)
;'''

# Комментарий к запросу: интересно, что без назначения dummy_alias выдает ошибку
# Видимо, без алиаса не может различить о каком publisher_id идет речь в подзапросе первого уровня вложения

output_3 = pd.io.sql.read_sql(query, con = engine)
output_3
```

## Запрос №4: Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками

```python
query = '''
WITH
multirated_books AS (SELECT book_id, COUNT(rating)
                     FROM ratings
                     GROUP BY book_id
                     HAVING COUNT(rating) >= 50  )

SELECT author AS top_rating_author
FROM authors
WHERE authors.author_id IN (SELECT author_id 
                            FROM (SELECT author_id, AVG(rating) AS avg_rating
                                  FROM books
                                  LEFT JOIN ratings ON books.book_id=ratings.book_id
                                  WHERE books.book_id IN (SELECT book_id FROM multirated_books)
                                  GROUP BY author_id
                                  ORDER BY avg_rating DESC
                                  LIMIT 1 ) AS dummy_alias )
;'''

# Комментарий к запросу: тут без join не обойтись, так как усредняется поле из одной таблицы с группировкой по полю другой

output_4 = pd.io.sql.read_sql(query, con = engine)
output_4
```

## Запрос №5 : Посчитайте среднее количество обзоров от пользователей, которые поставили больше 50 оценок

```python
query = '''
WITH
heavy_raters AS (SELECT username, COUNT(rating_id)
                 FROM ratings
                 GROUP BY username
                 HAVING COUNT(rating_id) > 50)
                 
SELECT AVG(nof_reviews) avg_nof_reviews_by_heavy_raters
FROM (SELECT username, COUNT(review_id) AS nof_reviews
      FROM reviews
      WHERE reviews.username IN (SELECT username FROM heavy_raters)
      GROUP BY reviews.username) AS dummy_alias
;'''

# Комментарий к запросу: 

output_5 = pd.io.sql.read_sql(query, con = engine)
output_5
```
