## Инфомодель объекта анализа

### Упрощенная модель обслуживания вызовов в колл-центре
К данной модели проводится привязка референсных точек, в которых определяются значения измеряемых параметров для расчета метрик эффективности.  Модель упрощена, поэтому часть данных может ей противоречить - см. список задач проекта по изучению объемов таких противоречий.  Итак, упрощенная модель предполагает:
1. Все поступающие из внешних сетей вызовы (параметры direction='in', internal=False) паркуются в общей очереди колл-центра, из которой большая часть попадает в очереди ожидания на обслуживание к конкретным операторам.
2. Все операторы равнозначны (нет специализации по контексту вызова или скилзам и опыту), при проключении на них входящих внешних вызовов система старается выбирать наименее занятого оператора (предположительно, самая короткая очередь или свободный оператор - но возможны и какие-то другие критерии маршрутизации)
3. Каждый вызов получает статус принятого (параметр is_missed_call=False) или пропущенного (параметр is_missed_call=True).  Включая вызовы, которые не были проключены на оператора, а также вызовы во внешнюю сеть колл-центрами обзвона.
4. Для обзванивающих колл-центров все исходящие внешние вызовы инициируются только операторами - нет автообзвона целевых абонентов.
5. Колл-центры четко делятся на два непересекающихся подмножества: колл-центры обзвона абонентов и колл-центры обслуживающие звонки абонентов по горячим номерам.  Характер работы и критерии эффективности операторов в каждом разные.
6. Внутренние вызовы (параметр internal=True) осуществляются одним оператором другому и фигурируют в собираемых данных дважды - один раз как исходящий (параметр direction='out') для оператора A, второй раз как входящий (параметр direction='in') для оператора B.

### Таблица семантики атрибутов типа вызова

| direction | internal | is_missed_call |   | что это за тип вызовов     | какой тип колл-центра |
|:---------:|:--------:|:--------------:|:-:|:---------------------------|:----------------------|
| 'in'  | True  | True  | | пропущенный внутренний звонок к оператору B  | любой |
| 'in'  | True  | False | | принятый внутренний звонок к оператору B     | любой |
| 'in'  | False | True  | | пропущенный звонок от абонента               | горячая линия | 
| 'in'  | False | False | | принятый звонок от абонента                  | горячая линия |
| 'out' | True  | True  | | пропущенный внутренний звонок от оператора A | любой |
| 'out' | True  | False | | принятый внутренний звонок от оператора A    | любой |
| 'out' | False | True  | | абонент не взял трубку при обзвоне           | целевой обзвон |
| 'out' | False | False | | абонент взял трубку при обзвоне              | целевой обзвон |

Здесь и далее оператор А инициирует внутренний вызов, а оператор В его принимает (или не принимает).

### Взаимосвязь цели конечного пользователя и критериев эффективности операторов, а также роль данного исследования в этом
Как мы видим, цель самих колл-центров в использовании предоставляемых им результатов для повышения утилизации операторов.  В принципе, почему бы напрямую не считать утилизацию рабочего времени операторов и ранжировать операторов по ней?  Однако правила мотивации персонала несколько иные - персоналу непосредственно не назначают бизнес-цели компании:
* Операторам назначаются KPI на языке их сферы ответственности.
* Задача технолога бизнес-процессов - изобрести такие KPI для всех участников производственного/операционного процесса, чтобы они, в ходе их выполнения, оптимизировали достижение целей компании.

Именно поэтому эффективность операторов проверяется на параметрах обслуживания внешних вызовов, а бизнес-целью компании является снижение расходов - то есть при некоторой нагрузке на колл-центр минимизировать число операторов в нем - за счет повышения их утилизации.

Роль данного исследования в первую очередь проверить на данных статистики - приводят ли указанные KPI в селекции операторов в пользу выполнения бизнес-цели.

**В этом плане сильный отход от предложенных критериев эффективности не имеет смысла, так как для работы логики взаимосвязи KPI операторов и бизнес цели компании-клиента необходимо, чтобы в период предоставленных данных операторы *уже работали бы в условиях доведенных до них KPI*!** 

Иными словами, если мы сейчас предложим еще некоторого "сферического коня в вакууме" в качестве критерия эффективности KPI, то прежде чем анализировать его влияние на бизнес-цели, нам надо будет довести его до операторов и потом собрать новые данные по их работе!

```python
# Теперь проверим применимость вышеописанной логики заполнения отсутствующих значений поля is_internal

maxlen = calls_df.groupby(['date','cc_id']).agg({'nof_calls':'count'})['nof_calls'].max() # количество зеркальных точно меньше 
counter_ops_mirrored = [0 for i in range(maxlen)] # будем подсчитывать, сколько встретится "зеркальных" записей
counter_cc_mirrored = [0 for i in range(maxlen)] # то же, но для случая с отсутствующим значением operator_id

# обрабатываем пропуски в is_internal для записей с присутствующим значением operator_id
for dt, id, incom, missed in (zip(
                                 missing_df[missing_df['operator_id'].notna()]['date'],
                                 missing_df[missing_df['operator_id'].notna()]['operator_id'],
                                 missing_df[missing_df['operator_id'].notna()]['is_incoming'],
                                 missing_df[missing_df['operator_id'].notna()]['is_missed']
                                           )
                                       ):
    same_date_id_2att_df = complete_df.query('date==@dt & operator_id==@id & is_incoming==@incom & is_missed==@missed')
    counter_ops_mirrored[len(same_date_id_2att_df)] += 1

print('\n Для половины случаев (где присутствует operator_id) пропуск в поле is_internal можно заполнить для',
     counter_ops_mirrored[1], 'записей из', sum(counter_ops_mirrored) )
print('              при этом противоречат нашей теории заполнения', sum(counter_ops_mirrored[2:]), 'записей,')
print('              а в', counter_ops_mirrored[0], 'случаях просто не нашлось зеркальных записей в исходных данных.')

# обрабатываем пропуски в is_internal для записей с отсутствующим значением operator_id
for dt, id, incom, missed in (zip(
                                 missing_df[missing_df['operator_id'].isna()]['date'],
                                 missing_df[missing_df['operator_id'].isna()]['cc_id'], # так как id оператора теперь нет
                                 missing_df[missing_df['operator_id'].isna()]['is_incoming'],
                                 missing_df[missing_df['operator_id'].isna()]['is_missed']
                                           )
                                       ):
    same_date_id_2att_df = complete_df.query('date==@dt & cc_id==@id & is_incoming==@incom & is_missed==@missed')
    counter_cc_mirrored[len(same_date_id_2att_df)] += 1

print('\n Для половины случаев (где отсутствует operator_id) пропуск в поле is_internal можно заполнить для',
     counter_cc_mirrored[1], 'записей из', sum(counter_cc_mirrored) )
print('              при этом противоречат нашей теории заполнения', sum(counter_cc_mirrored[2:]), 'записей,')
print('              а в', counter_cc_mirrored[0], 'случаях просто не нашлось зеркальных записей в исходных данных.')

# подводим итоги
print('______________________________________________________________________________________________________________')
print('\n ИТОГО: предложенным способом "заполнения по зеркальному" можно заполнить', 
      counter_ops_mirrored[1]+counter_cc_mirrored[1], 'записей из', sum(counter_ops_mirrored)+sum(counter_cc_mirrored),
      'пропусков в поле is_internal,' )
print(' в оставшихся', sum(counter_ops_mirrored)+sum(counter_cc_mirrored)-counter_ops_mirrored[1]-counter_cc_mirrored[1],
      'записях пропуски поля is_internal следует заполнить наиболее вероятным значением False.')
```

создадим соответствующие словари с ключем по комбинации атрибутов вызова:

#### Семантические словари для всех комбинаций атрибутов вызова (`is_incoming`,` is_internal`,` is_missed`) 

```python
lost_onIVR_dict = {
    (True,True,True)   :True,
    (True,True,False)  :False,
    (True,False,True)  :True,
    (True,False,False) :False,
    (False,True,True)  :True,
    (False,True,False) :False,
    (False,False,True) :False,
    (False,False,False):False
}
call_type_dict = {
    (True,True,True)   :'пропущенный внутренний звонок к оператору B',
    (True,True,False)  :'принятый внутренний звонок к оператору B',
    (True,False,True)  :'пропущенный звонок от абонента',
    (True,False,False) :'принятый звонок от абонента',
    (False,True,True)  :'пропущенный внутренний звонок от оператора A',
    (False,True,False) :'принятый внутренний звонок от оператора A',
    (False,False,True) :'абонент не взял трубку при обзвоне',
    (False,False,False):'абонент взял трубку при обзвоне'
}
cc_type_dict = {
    (True,True,True)   :'любой',
    (True,True,False)  :'любой',
    (True,False,True)  :'горячая линия',
    (True,False,False) :'горячая линия',
    (False,True,True)  :'любой',
    (False,True,False) :'любой',
    (False,False,True) :'целевой обзвон',
    (False,False,False):'целевой обзвон'
}
```

```python
# посмотрим сколько пропущенных записей приходятся на разные комбинации атрибутов вызова
# оформим как функцию:
def calls_pertype_table(df, show_ivr):
    calls_pertype_df = df.groupby(by=['is_incoming','is_internal','is_missed'], as_index=False).agg(
                                                {   'cc_id':'nunique',
                                                    'date':'nunique',
                                                    'nof_calls':'sum',
                                                    'net_duration':'sum',
                                                    'brut_duration':'sum'
                                                }).rename(columns={'cc_id':'cc_uq','date':'days_uq'}).join(
                        df.groupby(by=['is_incoming','is_internal','is_missed'], as_index=False).agg(
                                                {'date':'count'}).rename(columns={'date':'count'})['count'] )

    # так как количество записей будет самым разным, то более сравниваемый агрегируемый параметр будет по одному вызову 
    calls_pertype_df['one_talk_sec'] = round( calls_pertype_df['net_duration'] / calls_pertype_df['nof_calls'], 1) 

    # добавим классификацию вызовов по наборам атрибутов

    if show_ivr: 
        calls_pertype_df['lost_onIVR'] = calls_pertype_df.apply(
        lambda df: lost_onIVR_dict[(df['is_incoming'],df['is_internal'],df['is_missed'])], axis=1)

    calls_pertype_df['cc_type'] = calls_pertype_df.apply(
    lambda df: cc_type_dict[(df['is_incoming'],df['is_internal'],df['is_missed'])], axis=1)

    calls_pertype_df['call_type'] = calls_pertype_df.apply(
    lambda df: call_type_dict[(df['is_incoming'],df['is_internal'],df['is_missed'])], axis=1)

    return calls_pertype_df

calls_pertype_table(calls_df[calls_df['operator_id'].isna()], True)
```

## Целевое исследование критерия разделения колл-центров на две группы
Первоначальная идея довериться автоматическому критерию кластеризации колл-центров напрямую не может быть применена к текущему датасету, так как алгоритм не знает какие веса следует придавать разным координатам в матрице расстояний между элементами.  В проекте с предсказанием черна успех кластеризации был обусловлен тем, что сперва было проведено предсказание, из которого получили веса важности, а потом в автокластеризацию отобрали только те параметры, чьи веса были самыми существенными согласно тренировке предсказательной модели.  Тем самым алгоритму кластеризации задали приоритетные параметры именно для кластеризации по целевой переменной.

В нашем случае если провести автоматическую кластеризацию сейчас, то колл-центры рискуют быть разделенными по какому-нибудь условному "цвету ногтей операторов" сильнее, чем по их типу в отношении абонентов!  Поэтому невзирая на то, будем ли мы в итоге делать автоматическую кластеризацию, или в ручную, сперва необходимо вывести параметры различия колл-центров.   Хорошо бы ввести какой-то скоринг и по его диапазону определять тип.

### Скоринг типа колл-центра
Скоринг должен учитывать то, что все колл-центры разных размеров и нагрузки, а диапазон значений должен быть фиксирован одинаково для всех колл центров.

Начнем с добавления в датасет поля, которое будет принимать значения:
* 1 - если тип вызова возможен только для колл-центра обслуживающего горячие номера, на которые звонят абоненты;
* -1 - если тип вызова возможен только для колл-центра обслуживающего целевые кампании обзвона абонентов;
* 0 - если тип вызова возможен для любого типа колл-центра (из двух выше)

У нас уже был подобный словарь по типам вызовов - теперь его нужно немного изменить: 

```python
flag_dict = {
    (True,True,True)   :0,
    (True,True,False)  :0,
    (True,False,True)  :1,
    (True,False,False) :1,
    (False,True,True)  :0,
    (False,True,False) :0,
    (False,False,True) :-1,
    (False,False,False):-1
}

calls_df['flag'] = calls_df.apply(
    lambda df: flag_dict[(df['is_incoming'],df['is_internal'],df['is_missed'])], axis=1)
```

Если бы каждая запись соответствовала одному вызову, то скоринг колл-центров, средний по всему времени их присутствия в датасете, уже можно было бы посчитать группировкой по cc_id и агрегацией по среднему значению flag.  Но датасет содержит агрегированные вызовы - каждая запись имеет статвесом значение nof_calls.  Поэтому добавим еще одно поле:  

```python
calls_df['flag_w'] = calls_df['flag'] * calls_df['nof_calls'] # здесь w - от weighted
```

Теперь все готово в расчету общего скоринга колл-центров за все дни присутствия в датасете.  Заметим, что знак скоринга не поменяется от того, оставляем ли мы в расчете записи с flag=0, но чувствительность скоринга будет выше, если их не учитывать - так как нормируется на общее число записей и внутренние вызовы будут сильнее группировать скоринг в район нуля!  Но если не учитывать часть записей, то есть рис не получить в группировке и часть колл-центров - у которых только внутренние вызовы, поэтому в конце надо проверить, сколько колл-центров не получат скоринг (по логике он у них ровно 0).

```python
# итак в скоринге будет участвовать только срез записей, которые точно квалифицируются по типу колл-центра (flag != 0)
cc_df = calls_df.query('flag != 0').groupby('cc_id', as_index=False).agg({
                                              'nof_calls':'sum',
                                              'tariff':'first',
                                              'cc_lifetime':'max',
                                              'flag_w':'sum'
                                               })
cc_df['cc_score_total'] = cc_df['flag_w'] / cc_df['nof_calls']

cc_df.info()
display(cc_df.head())
print('Удалось присвоить скоринг', len(cc_df), 'колл-центрам из', calls_df['cc_id'].nunique(), 
      'колл-центров, по которым есть данные о вызовах')
```

### Анализ применимости типа колл-центра к оценке эффективности операторов

```python
# сравним доли разделяющихся по |score| > 0.5, как показатель эффективности разделения колл центров на типы нашим скорингом

tariff_df = cc_df.query('-0.5 <= cc_score_total <= 0.5').groupby('tariff').agg({'cc_id':'count'}).rename(
                                                                 columns={'cc_id':'|score| <= 0.5'})
tariff_df = tariff_df.join(
    cc_df.query('-0.5 > cc_score_total | cc_score_total > 0.5').groupby('tariff').agg({'cc_id':'count'}).rename(
                                                                 columns={'cc_id':'|score| > 0.5'})['|score| > 0.5']
)
tariff_df = tariff_df.join(
    cc_df.groupby('tariff').agg({'cc_id':'count'}).rename(columns={'cc_id':'nof_cc'})['nof_cc']
)

tariff_df.loc['all',:] = tariff_df.sum()
tariff_df = tariff_df.astype('int')

tariff_df['Доля смешанных колл-центров (%)'] = round(tariff_df['|score| <= 0.5']/tariff_df['nof_cc']*100,1)
tariff_df['Доля типовых колл-центров (%)'] = round(tariff_df['|score| > 0.5']/tariff_df['nof_cc']*100,1)

print('\n         ТАБЛИЦА УВЕРЕННОСТИ КВАЛИФИКАЦИИ ТИПА КОЛЛ-ЦЕНТРА')
display(tariff_df[['Доля смешанных колл-центров (%)', 'Доля типовых колл-центров (%)']])
```

### Поле `cc_id` и статистики полей `nof_calls`, `net_duration`, `brut_duration`, `сс_lifetime`   
Сами значения поля `cc_id` не имеют смысла - они нужны только для группировки операторов по колл-центрам.  Могут иметь какой-то смысл только распределения этого поля по смысловым образом агрегируемым величинам: 
* `operator_id   -` агрегатор nunique(): количество операторов в колл-центре;
* `nof_calls     -` агрегатор sum(): суммарное кол-во вызовов для данного колл-центра;
* `net_duration  -` агрегатор sum(): суммарное время общения операторов для данного колл-центра;
* `brut_duration -` агрегатор sum(): суммарное время ожидания+общения для данного колл-центра; (лучше выделить только ожидание)
* `tariff        -` агрегатор first(): тариф подписки колл-центра на "наши" услуги;
* `сс_lifetime   -` агрегатор max(): продолжительность пользования колл-центра "нашими" услугами;
* `cc_type       -` агрегатор first(): тип колл-центра по введенной нами классификации

Мы уже собирали похожий датасет, но в тот раз для увеличения чувствительности нашего метода скоринга мы делали срез данных.  Теперь повторим для всего основного датасета:

```python
all_cc_df = calls_df.groupby('cc_id', as_index=False).agg({
                                                           'operator_id':'nunique',
                                                           'nof_calls':'sum',
                                                           'net_duration':'sum',
                                                           'brut_duration':'sum',
                                                           'tariff':'first',
                                                           'cc_lifetime':'max',
                                                           'cc_type':'first'
                                               }).rename(columns={'operator_id':'nof_opers'})

# Часто удельные параметры помогают лучше сравнивать разномасштабные объекты, поэтому добавим их:

all_cc_df['wait_time'] = all_cc_df['brut_duration'] - all_cc_df['net_duration']
all_cc_df['avg_talk_time'] = all_cc_df['net_duration'] / all_cc_df['nof_calls']
all_cc_df['avg_wait_time'] = all_cc_df['wait_time'] / all_cc_df['nof_calls']
all_cc_df['avg_daily_calls_per_oper'] = all_cc_df['nof_calls'] / all_cc_df['nof_opers'] / all_cc_df['cc_lifetime']
```

## Целевое исследование метрик эффективности операторов

### Уточнение критериев эффективности
Согласно менеджерам продукта следует оценивать эффективность операторов по следующим критериям: 
1. много пропущенных входящих [внешних - *дополнено*] вызовов,
2. долгое ожидания ответа при входящих внешних [отвеченных - *дополнено*]вызовах, (дополнено, так как иначе дублируется с прошлым критерием - так как часть вызовов пропущенно после некоторого ожидания)
3. мало исходящих [внешних - *дополнено*] вывозов — в случае тех колл-центров, что специализируются на обзвонах, а не на обработке входящих звонков.

По первому критерию есть следующие моменты:
* как мы уже видели в главе 3.3.3.1, почти 99% всех пропущенных вызовов умирают на IVR не будучи проключенными на оператора.  Поэтому количество записей основного датасета оцениваемых по критерию 1 будет в 50 раз меньше чем таковых по критерию 2 (оба критерия для одного типа услуги - горячая линия).
* вызов может быть пропущен при коротком времени ожидания абонента, но также и при долгом, что означает разную степень вины оператора!

По второму критерию есть следующие моменты:
* чем больше будет вызовов в день, тем больше будет время ожидания - как мы видели на примере корреляционной матрицы в главе 3.4.  Это означает, что если применять данный критерий напрямую, то он выбросит в топ плохих просто тех, кто много работает!

Поэтому мы выдвигаем идею изменения критериев в **новый список**:
1. Критерий достаточной обеспеценности всего колл-центра операторами горячей линии - по доле потерянных внешних входящих вызовов, не проключенных на операторов. Именно по доле - так как в отличие от операторов колл-центры разномасштабны!
2. Критерий эффективности оператора по модели горячей линии - по среднему времени ожидания ответа (на один вызов)
3. Критерий эффективности оператора по модели горячей линии - по среднему времени ожидания ответа (на один вызов)

### Пороги по критериям
Для каждого из критериев есть приемлемые и неприемлемые диапазоны величин - это определяется нормой рынка услуг и не может быть изобретено аналитически.  У нас есть только основной датасет, это значит, что можно будет посмотреть их распределения и взять за порог уровень децимария неэффективных.

В отношении колл-центров можно сравнивать также среднюю продолжительность ожидания на IVR и среднюю продолжительность ожидания у операторов - если первая существенно превышает вторую, то это значит, что данному колл-центру не хватает операторов.

### Сложные моменты применения критериев и порогов:
1. Итак теперь для операторов критериев два, но оценивать общую эффективность нужно по каждому оператору - значит нужен принцип их замешивания в общий скоринг.
2. Как было обнаружено в целевом исследовании ранее эффективно разделить колл-центры не получится.  К тому же при этом из учета будет потеряна значительная часть активности операторов каждый день.  Вполне возможно, что кампании обзвона носят эпизодический характер и из всего периода работы может оказаться, что половину времени оператор поработал на дозвоне, а остальное на горячей линии - деление критериев эффективности по типу колл-центра приведет к неучету половины его работы!
3. Операторы работают не каждый день.  И их колл-центры тоже активны не все 4 месяца.
4. Ночная смена оператора (если такая есть) может быть разорвана на два дня статистики.

Ситуацию со сменами нам не разрешить в рамках имеющихся уже агрегированных по суткам данных - будем исходить из простой модели: оператор работает в день 8 часов, в каждый конкретный день оператор либо работает, либо отдыхает (посменно).  Фактор неежедневной работы операторов можно купировать рассмотрением средних от суточных статистик оператора.  Как мы помним, последние собираются по 8 вариантам вызовов:

```python
effectiveness_table = calls_pertype_table(calls_df.query('operator_id!=0'),False) # без IVR

criteria_dict = {
    (True,True,True)   :0,
    (True,True,False)  :0,
    (True,False,True)  :2,
    (True,False,False) :2,
    (False,True,True)  :0,
    (False,True,False) :0,
    (False,False,True) :3,
    (False,False,False):3
}

effectiveness_table['criteria_no'] = effectiveness_table.apply(
    lambda df: criteria_dict[(df['is_incoming'],df['is_internal'],df['is_missed'])], axis=1)

print('\n             СТАТИСТИКА ПО ТИПАМ ВЫЗОВОВ ПРОКЛЮЧЕННЫХ К ОПЕРАТОРАМ (без operator_id==0 <=> IVR)')
effectiveness_table.drop(columns=['cc_uq','days_uq','net_duration','brut_duration','one_talk_sec'])
```

```python
# добавим категорию критерия и к основному датасету для дальнейшего анализа:
def linef_criteria(df):
    if df['operator_id'] == 0: 
        if ~df['is_internal']: return 1
        else: return 0
    else: return criteria_dict[(df['is_incoming'],df['is_internal'],df['is_missed'])]

calls_df['criteria_no'] = calls_df.apply(linef_criteria, axis=1)

print('\n Всего за период предоставленных данных работали', calls_df['operator_id'].nunique()-1, 'операторов,')

print('   из них', calls_df.query('criteria_no in (2,3)')['operator_id'].nunique(), 'операторов обрабатывали и внешние вызовы,')
print('                 отработав при этом', calls_df.query('criteria_no in (2,3)')['operator_id'].count(), 'человеко-дня;')

print('   из последних', calls_df.query('criteria_no in (2,3)').groupby('operator_id', as_index=False).agg(
                                          {'criteria_no':'nunique'}).query('criteria_no==1')['operator_id'].nunique(), 
      'операторов работали только в одном единственном режиме все это время.')
print('                 отработав при этом', calls_df.query('criteria_no in (2,3)').groupby('operator_id', as_index=False).agg(
                                          {'criteria_no':'nunique','date':'count'}
                                          ).query('criteria_no==1')['date'].sum(), 'человеко-дня.')
```

#### Описание идеи метода по построению скоринга операторов
1. Для каждого оператора будем оценивать среднее значение по рабочим дням для обоих критериев за дни работы данного оператора и сравнивать с порогами децимария по дневным нормам выработки, которые определим по статистике соответствующих полей.
2. Поскольку мы предполагаем, что любой оператор может работать мало или совсем не работать по какому-то критерию, нам нужно учитывать также некоторый вес данного критерия.
3. В итоге у нас будет суд из двух присяжных в отношении каждого оператора, причем мнение этих присяжных будет учитываться пропорционально весам соответствующих услуг в общей утилизации оператора в его рабочие дни.

Попробуем далее собрать датасет в соответствии с вышеописанной методологией

### Сборка датасета  `daily_op_score_df`
Как мы убедились выше, основной датасет недостаточен для статистического анализа различий между операторами - так как ID оператора не является значимой величиной (кроме 0, что означает IVR, который будет исследован в отдельном целевом исследовании).  Поэтому нужно собрать более подходящий датасет по операторам:

```python
# оставляем только оцениваемые вызовы операторам
scored_calls_df = calls_df.query('criteria_no in (2,3)').sort_values(['operator_id','date']).reset_index(drop=True)

# там есть те, у кого в один день записи с разным типом оценки, для сохранения поэтому вводим вес критерия
scored_calls_df['daily_scored_calls'] = scored_calls_df.groupby(['operator_id','date'])['nof_calls'].transform(sum)
scored_calls_df['weight'] = scored_calls_df['nof_calls'] / scored_calls_df['daily_scored_calls']

# эти поля не нужны
no_show_cols = ['is_incoming','is_internal','is_missed','cc_type','daily_scored_calls','cc_lifetime']
scored_calls_df.drop(columns=no_show_cols).head(10)
```

Все почти готово - однако есть проблема, мешающая замешивать критерии, - у нас по два типа вызовов вносят свой вклад в скоринг по каждому критерию, поэтому их вклад один день нужно сложить для каждого оператора:

```python
daily_op_score_df = scored_calls_df.groupby(['operator_id','date','criteria_no'], as_index=False).agg({
                                            'weight':'sum',
                                            'nof_calls':'sum',
                                            'net_duration':'sum',
                                            'brut_duration':'sum',
                                            'cc_id':'first',
                                            'tariff':'first'
})

# теперь можно заполнить столбец со скорингом - NB!: он разнотипный по полю criteria_no
def funcf_score(df):
    if df['criteria_no'] == 2: return ( df['brut_duration'] - df['net_duration'] ) / df['nof_calls']
    elif df['criteria_no'] == 3: return df['nof_calls']
    else: return 0
    
daily_op_score_df['score'] = daily_op_score_df.apply(funcf_score , axis=1)
```

все отлично, теперь определим пороги статистикой поля score отдельно по каждому из критериев.

### Определение смешанного скоринга операторов
Итак, по порогам решение принято, далее есть два варианта сборки смешенного скоринга:
1. Для каждого оператора определяем среднее по дням его работы значение оцениваемого параметра, сравниваем его с порогом по критерию данного параметра и выносим решение по эффективности по каждому критерию: d2 и d3, соответственно, со значением True - если неэффективен, для каждого.  Далее используем суммарные веса, чтобы вычислить смешанный скоринг:
$$ score_{mixed} = {W_2*d_2 + W_3*d_3}\over{W_2 + W_3}$$ где
$$ W_2 = \sum_{по\ всем\ 2}{weight}, \ \ \ \ \ \ \ W_3 = \sum_{по\ всем\ 3}{weight}$$
2. Для каждого рабочего дня выносим решение по эффективности по критериям d2 и d3, и вычисляем ежедневный смешанный скоринг как
$w_2*d_2 + w_3*d_3$, где $w_i=weight$ для записи с критерием оценки $i$, после чего смешанный скоринг определяется усреднением по всем рабочим дням оператора дневного смешанного скоринга:
$$ score_{mixed} =\ <w_2*d_2 + w_3*d_3>_{по\ всем\ рабочим\ дням\ оператора}$$
Во втором случае уже не нужно делить на сумму $w_i$ так как в рамках одного дня для данного оператора она всегда равна 1 по правилу сборки поля weight.

Нам больше нравится второй способ, так как в нем нельзя выполнить стахановский подвиг в один день, а потом отдыхать месяц - все равно средний уровень показателя будет в хорошем диапазоне.  Предполагается, что хорошо работать нужно каждый день.  Поэтому используем второй способ расчета смешанного скоринга.

```python
# выбираем пороги единые для всех операторов и колл-центров
thres2 = 35.0
thres3 = 9.6

def linef_decision(df):
    if df['criteria_no'] == 2:
        if df['score'] > thres2: return True # децимарий по горячей линии
        else: return False
    elif df['criteria_no'] == 3:
        if df['score'] < thres3: return True # отсев по целевому дозвону 
        else: return False
    else: 
        print('something wrong!')
        return True

daily_op_score_df['decision'] = daily_op_score_df.apply(linef_decision, axis=1)
daily_op_score_df['w_times_d'] = daily_op_score_df['weight'] * daily_op_score_df['decision']

# теперь нужно агрегировать датасет по суткам сложив поле w_times_d по двум критериям для каждого дня:
daily_scored_op_df = daily_op_score_df.groupby(['operator_id','date'], as_index=False).agg({
                                              'w_times_d':'sum',
                                              'nof_calls':'sum',
                                              'net_duration':'sum',
                                              'brut_duration':'sum',
                                              'cc_id':'first',
                                              'tariff':'first'
}).rename(columns={'w_times_d':'mixed_score'})

# осталось агрегировать датасет по операторам усреднив поле mixed_score по рабочим дням для каждого оператора:
op_score_df = daily_scored_op_df.groupby(['operator_id'], as_index=False).agg({
                                              'mixed_score':'mean',
                                              'date':'nunique',
                                              'nof_calls':'sum',
                                              'net_duration':'sum',
                                              'brut_duration':'sum',
                                              'cc_id':'first',
                                              'tariff':'first'
}).rename(columns={'date':'nof_days_worked'}).sort_values(by=['mixed_score','nof_calls'],
          ascending=[False,True]).reset_index(drop=True)                                  # кандидаты на увольнение вперед!

# добавим также еще одно поле к датасету скоринга операторов, отражающее бизнес-цель повышения утилизации операторов:
op_score_df['avg_util'] = round(op_score_df['net_duration'] / op_score_df['nof_days_worked'] / (8*60*60) *100, 1)
```

## Исследование статистики непроключенных вызовов - вызовы застрявшие на IVR
Вызовы не проключенные на оператора не попали в сферу его ответственности и поэтому не могут влиять на персональный скоринг операторов.  Тем не менее, они характеризуют качество услуг всего колл-центра в целом с точки зрения абонентов и могут служить косвенным симптомом нехватки операторов.  В пользу последнего говорит обнаруженный в главе 3.3.3.1 факт того, что долее половины внешних входящих вызовов остаются неотвеченными еще до проключения на операторов. 

```python
ivr_calls_df = calls_df.query('operator_id==0').reset_index(drop=True) # исследуем статистику данного датасета

ivr_calls_df.info()
print('_________________________________________________________________________________________________________')
ivr_calls_df.head()
```

```python
print('Потерянные до проключения на операторов вызовы имеет', ivr_calls_df['cc_id'].nunique(), 
      'колл-центров из', calls_df['cc_id'].nunique() )
```

```python
# статистику остальных полей посмотрим, собрав таблицу по типам вызовов
# атрибуты вызова при этом сохранять не нужно, так как каждой строке соответствует их уникальная семантическая комбинация 

ivr_calls_df['call_type'] = ivr_calls_df.apply(
    lambda df: call_type_dict[(df['is_incoming'],df['is_internal'],df['is_missed'])], axis=1)

ivr_stats_df = ivr_calls_df.groupby('call_type').agg({
                                                       'cc_id':'nunique',
                                                       'nof_calls':'sum',
                                                       'net_duration':'sum',
                                                       'brut_duration':'sum',
                                                       'tariff':'nunique',
                                                       'cc_type':'nunique'
    
}).rename(columns={'cc_id':'nof_cc','tariff':'nof_tariffs', 'cc_type':'nof_cc_types'}).sort_values('nof_cc', ascending=False
)

ivr_stats_df.index = ['потерянный звонок абонента',
                      'потерянная входящая нога внутреннего вызова',
                      'потерянная исходящая нога внутреннего вызова']

ivr_stats_df['avg_wait_time'] = round( ivr_stats_df['brut_duration'] / ivr_stats_df['nof_calls'] , 1)

ivr_stats_df
```

Содержательной для нашего анализа является только первая строка - внешние звонки от абонента, потерянные до проключения оператору.  Посмотрим статистику соответствующего среза по колл-центрам - сравним эффективность IVR и недостаток операторов в них.

### Сбор и исследование датасета cc_ivr_df

```python
cc_ivr_df = ivr_calls_df.query('is_incoming & (~is_internal) & is_missed').groupby('cc_id').agg({
                                                                                   'nof_calls':'sum',
                                                                                   'net_duration':'sum',
                                                                                   'brut_duration':'sum',
                                                                                   'tariff':'first',
                                                                                   'cc_type':'first'
}).rename(columns={'nof_calls':'nof_ivr_lost_calls'})
cc_ivr_df['avg_wait_time'] = round( cc_ivr_df['brut_duration'] / cc_ivr_df['nof_ivr_lost_calls'] , 1)

cc_ivr_df = cc_ivr_df.join(calls_df.groupby('cc_id').agg({'operator_id':'nunique'}).rename(
                           columns={'operator_id':'nof_ops'}) )

cc_ivr_df['ivr_lost_calls_per_op'] = round( cc_ivr_df['nof_ivr_lost_calls'] / cc_ivr_df['nof_ops'] , 1)

cc_ivr_df = cc_ivr_df.join(calls_df.query('criteria_no in (2,3)').groupby('cc_id').agg({'nof_calls':'sum'}).rename(
                           columns={'nof_calls':'nof_served_calls'}) ) # в этом поле могут быть пропуски
# пропуски реально соответствуют нулевым значениям этого поля - заменим их
cc_ivr_df['nof_served_calls'] = cc_ivr_df['nof_served_calls'].fillna(0).astype('int64')

cc_ivr_df['served_vs_lost'] = round( cc_ivr_df['nof_served_calls'] / cc_ivr_df['nof_ivr_lost_calls'] , 2)

cc_ivr_df = cc_ivr_df.sort_values(by=['ivr_lost_calls_per_op','avg_wait_time'], ascending=[False,False])

cc_ivr_df.info()
print('______________________________________________________________________________________________________')
cc_ivr_df
```

* Далее выводы будут программными - так как в них будут рассчитываемые величины:

```python
# Вводим порог по первому критерию для IVR:
thres1 = 11.11   # соответствует уровню 10% потерь входящих вызовов от абонентов на IVR

print('\n ВНИМАНИЕ НА ОСНОВНУЮ ПРОБЛЕМУ БОЛЬШИНСТВА КОЛЛ-ЦЕНТРОВ: \n')

print( len(cc_ivr_df.query('served_vs_lost < @thres1')),'(', 
       round(len(cc_ivr_df.query('served_vs_lost < @thres1')) / len(cc_ivr_df) * 100,1),
      '% от всех) колл-центров теряет более 10% вызовов абонентов горячей линии на IVR')

print( len(cc_ivr_df.query('served_vs_lost < 1')),'(', 
       round(len(cc_ivr_df.query('served_vs_lost < 1')) / len(cc_ivr_df) * 100,1),
      '% от всех) колл-центров теряет более 50% вызовов абонентов горячей линии на IVR')

print( len(cc_ivr_df.query('avg_wait_time > @thres2')),'(', 
       round(len(cc_ivr_df.query('avg_wait_time > @thres2')) / len(cc_ivr_df) * 100,1),
      '% от всех) колл-центрам просто не хватает операторов!')
```

# Проверка гипотез 

## Гипотеза 1: предложенные продуктовым маркетингом мотивационные KPI эффективности операторов приводят к повышению утилизации операторов

```python
# готовим сравниваемые выборки
sample_1 = list(op_score_df.query('mixed_score ==1')['avg_util']) # неэффективные операторы
sample_2 = list(op_score_df.query('mixed_score < 1')['avg_util']) # более эффективные операторы

p_value = stats.ttest_ind(sample_1, sample_2, equal_var = False).pvalue

print('Средняя утилизация неэффективных операторов', round(mean(sample_1),2),
      '% отличается от средней утилизации остальных операторов', round(mean(sample_2),2),
      '%\n        со статистической значимостью разницы средних на уровне p-value =', p_value)
```

### Выводы по гипотезе №1:
* Мы можем подтвердить с любой разумной статистической достоверностью (p-value = $1.8 * 10^{-38}$), что утилизация рабочего времени 15.5% неэффективных операторов, отобранных нашим методом, на порядок меньше утилизации остальных операторов, не попавших в черный список.
* Данный успех в подтверждении гипотезы является доказательством эффективности как нашего метода скоринга эффективности операторов, так и выбора KPI для мотивации операторов - в плане его соответствия бизнес-цели клиента по снижению расходов на зарплаты операторам (за счет увеличения утилизации их рабочего времени).
