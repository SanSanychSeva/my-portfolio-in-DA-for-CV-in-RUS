### Участники обоих тестов
Использовать для проверки каждого вопроса срезы объединенного датасета участников неудобно, поэтому создадим переменную-набор всех множеств участников тестов в отношении их разбиения между тестами и группами - так, чтобы их него можно было индексацией извлекать любой вариант множества тестируемых пользователей.

#### Датасет множеств участников тестов по комбинации (тест,группа) - `cobaye_sets_df`

(**cobaye** [кобай], *фр.* - подопытное животное/свинка/кролик)

```python
# составим таблицы тестируемых пользователей для всех 4-х комбинаций (тест,группа):

cobaye_sets_df = pd.DataFrame( [
    [
        set(enrolled_df.query(' group == "A" & ab_test == "interface_eu_test" ')['user_id']),
        set(enrolled_df.query(' group == "B" & ab_test == "interface_eu_test" ')['user_id'])
    ],
    [
        set(enrolled_df.query(' group == "A" & ab_test == "recommender_system_test" ')['user_id']),
        set(enrolled_df.query(' group == "B" & ab_test == "recommender_system_test" ')['user_id'])
    ]
],
    index=['I-test','R-test'],
    columns=['A','B']
)
```

Собственно, это было нужно, чтобы далее легко проверять тестируемых пользователей, одновременно участвующих в различных тестах и группах, используя элементарную алгебру множеств - всего одним оператором **&**:

```python
# например, посмотрим сколько тестируемых одновременно в группе В теста интерфейса и в группах на тесте системы рекомендаций:

print('На нашем тесте в группе А есть',len( cobaye_sets_df.loc['R-test','A'] & cobaye_sets_df.loc['I-test','B']), 
      'тестируемых пользователей из группы В параллельного теста')
print('На нашем тесте в группе B есть',len( cobaye_sets_df.loc['R-test','B'] & cobaye_sets_df.loc['I-test','B']), 
      'тестируемых пользователей из группы В параллельного теста')
```

Однако перебирать построчно все варианты неудобно, да и можно упустить какой-нибудь - создадим таблицу взаимных пересечений, чтобы ничего не упустить:

#### Датасет фликеров всех вариантов (тест,группа) х (тест,группа ) - `show_flickers_df`

```python
# выведем все возможные варианта двойного участия тестируемых пользователей в разных тестах и группах:
# (FYI: попытка работать с индексами-кортеджами напрямую не получилась - пришлось значения индексов вывести в отдельные поля)

testXgroup = [ (t,gr) for t in ('I-test','R-test') for gr in ('A','B') ] # вспомогательный список всех сочетаний (тест, группа)

show_flickers_df = pd.DataFrame(
    index=[i for i in range(len(testXgroup))],
    columns=[i for i in range(len(testXgroup))]
)
show_flickers_df['crossing with'] = testXgroup
show_flickers_df_T = show_flickers_df.T
show_flickers_df_T['crossing with'] = testXgroup+['none']
show_flickers_df = show_flickers_df_T.T

for i in range(len(testXgroup)):
    for c in range(len(testXgroup)):
        if c <= i: show_flickers_df.loc[i,c] = len(cobaye_sets_df.loc[testXgroup[i][0],testXgroup[i][1]] & 
                                                  cobaye_sets_df.loc[testXgroup[c][0],testXgroup[c][1]] )

print('\n КОЛИЧЕСТВА ТЕСТИРУЕМЫХ, ОДНОВРЕМЕННО УЧАСТВУЮЩИХ В НЕСКОЛЬКИХ ТЕСТАХ И/ИЛИ ГРУППАХ')
show_flickers_df # как как таблица симметрична, то условие c<=i - чтобы не отвлекаться на повторяющиеся значения 
# при этом диагональные значения дают количества пользователей для каждого набора (тест,группа)
```

#### Выводы по участникам обоих тестов:
* Заметим, что **реальных фликеров (сменивших группу в ходе одного теста) нет ни в одном из тестов** - два нуля в таблице!  
* Зато **и в группе А и в группе В нашего теста есть тестируемые из обеих групп параллельного теста** - до 1/4 от всей группы нашего теста!
* С одной стороны, в тестах используется теория возмущений, согласно которой, даже если воздействия тестов не полностью ортогональны (то есть взаимодействие есть), но из-за малости возмущения квадраличным членом взаимодействия можно пренебречь по сравнению с линейным, а в линейном приближении все тесты независимы: если уровень линейного эффекта от тестируемого фактора ожидается в обоих тестах на уровне 10%, то квадратичное взаимодействие будет на уровне всего лишь 10% * 10% = 1% ! 
* С другой стороны, возможны и дополнительные доводы в отношении интерференции тестов у участников обоих:  
 * Даже если таковая и существенна, но если соотношение участников параллельного теста в обеих группах нашего теста пропорционально таковому для всех участников нашего теста, то их влияние будет пропорциональным на обе группы нашего теста и не будет видно в величинах представляющих относительное изменение статистик между группами.
 * Кроме того, тесты могут планироваться в одно время, но проводиться в разное - следует еще попытаться сравнить наложение временных периодов проведения тестов.
    
### Сравнение пропорциональности замеса пользователей из параллельного теста в наш тест
Поскольку мы не знаем, какая из групп параллельного теста может оказать большее влияние (по исходному состоянию групп А для обеих, теоретически сильнее влияет группа В, но мы не знаем какие из параллельно тестируемых изменений пойдет в пром) - поэтому проверим их отдельно

```python
# используем наш чрезвычайно мощный инструмент анализа множеств:

show_df = show_flickers_df.fillna(0)
show_df['total'] = show_df[2] + show_df[3]
show_df = show_df.drop(columns=['crossing with']).T.drop(columns=[0, 1, 'crossing with'])
show_df['total'] = show_df[2] + show_df[3]

show_df[2] = (show_df[2] / show_df['total'] * 100).apply(lambda n: round(n,1))
show_df[3] = (show_df[3] / show_df['total'] * 100).apply(lambda n: round(n,1))
show_df = show_df.T.drop(columns=[2, 3])

show_df.index = ['Тестируемых в группе A теста R (в % по R)','Тестируемых в группе B теста R (в % по R)',
                 'Тестируемых в тесте R (соответствует 100%)']
show_df.columns = ['при этом они есть в группе A теста I','при этом они есть в группе B теста I','всего тестируемых']
show_df['units'] = ['%','%','чел.']
print('\n       СРАВНИТЕЛЬНАЯ ТАБЛИЦА СООТНОШЕНИЙ РАЗНОТИПНЫХ ФЛИКЕРОВ ИЗ ПАРАЛЛЕЛЬНОГО ТЕСТА В ТЕСТЕ СИСТЕМЫ РЕКОМЕНДАЦИЙ')
show_df
```

```python
show_df = show_df.iloc[[0,1],:].reset_index(drop=True)
show_df['при этом они есть в группе A теста I'] = (show_df['при этом они есть в группе A теста I'] -
                                                   show_df['всего тестируемых'])
show_df['при этом они есть в группе B теста I'] = (show_df['при этом они есть в группе B теста I'] -
                                                   show_df['всего тестируемых'])
show_df = show_df.drop(columns=['всего тестируемых','units'])
show_df.columns = ['разница в делении для теста R с группой A теста I',
                   'разница в делении для теста R с группой B теста I',]
show_df.index = ['для группы A теста R','для группы B теста R']
show_df
```

### Сравнение сроков конкурирующих тестов
К сожалению у нас не дан период проведения второго теста, поэтому мы можем пытьться оценить его перекрытие с нашим тестом путем просмотра событий для тестируемых пользователей, которые есть в параллельном тесте интерфейса, но отсутствуют в нашем тесте системы рекомендаций.  Воспользуемся алгеброй множеств: 

```python
# возьмем множество user_id, которые есть в параллельном тесте, но отсутствуют в нашем:
mono_cobaye_I_test = ( (cobaye_sets_df.loc['I-test','A'] | cobaye_sets_df.loc['I-test','B']) - 
                       (cobaye_sets_df.loc['R-test','A'] | cobaye_sets_df.loc['R-test','B']) )
# а заодно и симметричное ему:
mono_cobaye_R_test = ( (cobaye_sets_df.loc['R-test','A'] | cobaye_sets_df.loc['R-test','B']) - 
                       (cobaye_sets_df.loc['I-test','A'] | cobaye_sets_df.loc['I-test','B']) )
```

```python
# и теперь посмотрим, как распределены разные типы событий для каждого из тестов (только по однотестовым участникам)

plt.figure(figsize=(20, 12))
myax = [plt.subplot(2,2,i+1) for i in range(4)]

event_list = ['login', 'product_page', 'product_cart', 'purchase']
logs_df['date'] = logs_df['event_date'].apply(lambda dt: dt.day) # а то даты целиком не помещаются - выводим только число

for i in range(4):
    sns.countplot(data=logs_df.query('(user_id in @mono_cobaye_I_test) & (event_name == @event_list[@i])'), 
                  ax=myax[i], x='date', label='только из параллельного теста', color='red', alpha=0.5)
    sns.countplot(data=logs_df.query('(user_id in @mono_cobaye_R_test) & (event_name == @event_list[@i])'), 
                  ax=myax[i], x='date', label='только из нашего теста', color='green', alpha=0.5)
    
    myax[i].set_title('Распределение события ' + event_list[i] + ' по дням - для монотестовых участников')
    myax[i].set_xlabel('дата, начиная с 7 по 30 декабря 2020г.')
    myax[i].set_ylabel('суточное количество событий')
    myax[i].set_xticks(ticks=[i for i in range(24)])
    myax[i].set_xlim(right=24)
    myax[i].grid()
    myax[i].legend()

logs_df = logs_df.drop(columns=['date'])
print('\n                     РАСПРЕДЕЛЕНИЕ РАЗНЫХ ТИПОВ СОБЫТИЙ ДЛЯ УЧАСТНИКОВ ТОЛЬКО ОДНОГО ИЗ ТЕСТОВ')
plt.show()
```

```python
# это блок проверок ТЗ
def check_df_vs_TRS(df):
    nof_TRS_errors = 0
    print('проводим тест финального датасета на соответствие ТЗ:')

    if df['user_id'].nunique() < 6000:
        print('* Недостаточно активных участников в нашем тесте')
        nof_TRS_errors += 1

    if (df['event_date'].min() < datetime(2020,12,7) ) or (df['event_date'].max() > datetime(2021,1,4) ):
        print('* Дата события вне временных рамок нашего теста')
        nof_TRS_errors += 1
    
    if len(df.query('ab_test != "recommender_system_test"') ) > 0:
        print('* Есть события не из нашей воронки')
        nof_TRS_errors += 1
    
    if (df['first_date'].min() < datetime(2020,12,7) ) or (df['first_date'].max() > datetime(2021,1,4) ):
        print('* Есть участники не из нашего периода набора новых')
        nof_TRS_errors += 1
    
    if len(df.query('region != "EU"') ) > 0:
        print('* Есть участники не из EU')
        nof_TRS_errors += 1

    print('По результатам проверки ТЗ выявлено', nof_TRS_errors, 'типов несоответствия ТЗ', end='')
    if nof_TRS_errors > 0: print(' - список выше')
    
check_df_vs_TRS(full_logs_df)
```

# Результаты A/B-тестирования 

## подготовка выборок для z-критерия
для изучения статвеличин типа бинарного выбора 0-1 при определении доли 1 нужен датасет по участникам, а не по событиям - соберем его:

```python
test_logs_df['name'] = test_logs_df['event_name'].apply(lambda s: s+',')
test_users_df = test_logs_df.groupby('user_id', as_index=False).agg(
                                   {'group':'first', 'name':'sum'}).rename(columns={'name':'names_list'})
test_users_df['names_list'] = test_users_df['names_list'].apply(lambda s: s.split(','))

for e in event_list:
    test_users_df[e] = test_users_df['names_list'].apply(lambda ls: int(e in ls))    

# поскольку конверсию будем брать с этапа логина
test_vs_login_df = test_users_df.query('login==1').reset_index(drop=True) 

test_vs_login_df.info()
print('_'*50)
test_vs_login_df.head()
```

**Инструкция по получению выборки для z-критерия**: нужно сделать срез `test_vs_login_df` по нужной группе и выбрать столбец с соответствующим именем этапа воронки.  Две группы - две выборки для сравнения в z-критерии.

## Проверка статистической значимости долей z-критерием.

### Нулевая и альтернативная гипотезы
так как нам нужно узнать не просто, что эффект, а что он превышает ожидаемый уровень, то нулевая гипотеза будет уже не о равенстве:
* $H_0$ - среднее в группе В равно среднему в группе А, увеличенному на 10%;
* $H_1$ - среднее в группе В превышает среднее в группе А, увеличенное на 10%.

статистика для проверки z-критерия задается формулой статвеличины с нормальным распределением $N(0,1)$:
$$ {(p_1 - p_2 - \Delta_{expected})}\over{\sqrt{p\ (1-p)(1/n_1+1/n_2)}} $$

### Проверка статзначимости для эффекта на CR из login в purchase
наши выборки в поле `purchase` датасета `test_vs_login_df` - в группах А и В соответственно

```python
def get_z_citeria(event,test_df,effect=0.1):
        
    p1 = test_df.query('group == "B"')[event].mean()
    p2 = test_df.query('group == "A"')[event].mean()
    delta = effect*p2 # тут мы вносим ошибку следующего порядка, так как p2 - не равно истинному параметру в популяции
    p = test_df[event].mean()
    n1 = test_df.query('group == "B"')[event].count()
    n2 = test_df.query('group == "A"')[event].count()

    z_value = (p1 - p2 - delta) / math.sqrt( p * (1-p) * (1/n1 + 1/n2) ) 
    p_value = (1 - stats.norm(0, 1).cdf(z_value))      # смотрим односторонний критерий - поэтому на 2 не умножаем

    print('\n Z-статистика для конверсии в', event, ':')
    print('-'*50)
    print('Доля конверсии в группе В', round(p1*100,3), '%')
    print('Доля конверсии в группе А', round(p2*100,3), '%')
    print('величина эффекта в пользу В', round((p1-p2)/p2 *100,3), '%')
    print('p-value такой разницы между полученным эффектом и ожидаемым (', round(effect*100,2),
          '%) составил', round(p_value, 3))

get_z_citeria('purchase',test_vs_login_df,0.0)
print('_'*50)
get_z_citeria('purchase',test_vs_login_df)
print('_'*50)
get_z_citeria('purchase',test_vs_login_df,0.118)
print('_'*50)
print('\n Для сравнения ниже - конверсия, посчитанная от регистрации, а не от логина:', end=' ')
get_z_citeria('purchase',test_users_df,0.0)
print('_'*50)
get_z_citeria('purchase',test_users_df)
```

### Заключение по статзначимости и эффекту на CR из login в purchase
* Заметим, что величина эффекта несколько иная, чем была в таблице ДЕТАЛЬНЫЕ ХАРАКТЕРИСТИКИ ВОРОНКИ ПРОДАЖ И ИХ ОТНОСИТЕЛЬНЫЕ ИЗМЕНЕНИЯ ОТ ГРУППЫ А К В: там было 15.39%, а здесь только 14.38% - потярян 1% в эффекте, причина в том, что там конверсия считалась делением числа покупок на число логинов, а здесь оставлены только пользователи, у которых был логин - то есть были активные участники, которые совершили покупку прямо из состояния регистрации без логинов (после данных выводов мы поищем таких для подтверждения).
* Тем не менее, полученная в тестах конверсия 14.4% значительно превышает ожидаемые 10% - даже с гарантированным запасом из-за маржи безопасности на пропорциональность замеса участников из параллельного теста.  Его мы оценивали в 11.8%.
