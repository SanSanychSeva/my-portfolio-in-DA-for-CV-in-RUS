## Бизнес-определение оттока в данном проекте:
Показатель оттока - есть показатель бизнес-результатов календарного месяца. 
Считается, что Клиент попал в отток по результатам месяца **М**, если: 
1. Клиент не посещал фитнес-центр весь данный месяц **М**.
2. На начало данного месяца Клиент был активным.
 
Клиент считается активным на начало месяца **М**, если в предыдущем календарном месяце он хотя бы раз посетил фитнес-центр.

Собственно Оттоком (`churn`) в месяце **М** называется доля Клиентов попавших в отток в данном месяце от количества активных Клиентов на начало данного месяца. 

Таким образом Отток (`churn`) равен вероятности для одного активного Клиента ни разу не посетить фитнес центр за один календарный месяц.

### Обнаружены системные ошибки в датасете!

1. Данные 210 записей системно противоречат нашему исследованию - они не соответствуют совокупности:
 * разъяснений семантики полей датасета, полученных от его источников,
 * бизнес-определения оттока, заданного при определении целей исследования данного датасета.
2. Так, например, запись номер 0 непонятна - если уже был зафиксирован факт отсутствия посещений в прошлом месяце, то почему статус оттока до сих пор 0?
3. Самыми парадоксальными являются записи с нулевой частотой посещений, и при этом со значением churn==0 (False): например, это номера записей 31 и 3983.  Забавно, что у обеих записей есть дополнительные расходы - посещений нет, а расходы есть.  Возможно человек записался только ради массажа, а в зал не заходит, поэтому имеет 0 посещений, которые регистрируются при выдаче ключа от шкафчика и полотенца.  
4. Запись номер 13 также удивительна - хотя статус чернера соответствует частоте посещений, но у нас есть дополнительный расход.  Видимо, часть услуг можно получать и без посещений, а тогда нужно менять определение оттока и добавлять параметры - частоты пользования услугами без посещения. 
5. Записи подобные номерам 21, 46, 3957, 3966, 3971 и 3977 "наиболее нормальные" - видимо, их источник просто невнимательно прочел определение оттока - эти клиенты должны были попасть в отток еще в прошлый месяц, так как уже в прошлый месяц они ни разу не были в фитнес-центре.  

#### Идеи по обработке системных ошибок:
1. Все записи, у которых значение поля частоты посещений в последний месяц равны 0.0 следует удалить, потому что:
 * если при этом churn==0, то это даже не интерпретируется в рамках целей проекта;
 * если при этом churn==1, то значения последнего месяца все равно утеряны и такие записи будут только искажать предсказательную модель.
2. Посмотрим на записи, у которых нулевым является только значение средней частоты посещений, но присутствует ненулевая посещаемость за последний месяц: 

```python
anom = gym_df.query('avg_class_frequency_total==0.0 & avg_class_frequency_current_month!=0.0').rename(columns=gym_col_dict)
print('\n Количество записей с tot_freq==0.0 и mo_freq!=0.0 равно', f'{len(anom)}, а именно:')
anom
```

3. По нашему мнению, сохранив значимость частоты последнего месяца для предсказательной модели оттока, "спасти" можно только записи, у которых значение lifetime (mo_pass) не больше 1 - при этом среднюю частоту можно взять равной частоте за последний месяц.  Но таких записей всего 6 на фоне 210 бракованных записей - такая разница не стоит усилий по их спасению!
4. Поэтому решение принимается удалить все 210 записей, что составит 5,25% исходного датасета - это много, но данные записи явно не соответствуют предполагаемой семантике полей датасета и бизнес-определению оттока, поэтому являются бракованными с точки зрения целей проекта.  (См. также далее главу 2.3.2.1 - с ярким примером того, какие аномалии распределений параметров возникают, если не удалить бракованные записи)
5. NB!: После их удаления придется повторно прогнать всю главу статистики полей. 
6. Можно предположить, что определение оттока не было четко дано при ведении данных и/или при их сборке - в дальнейшем это является замечанием к исходным данным.  По факту, нули в значении частот мы трактуем как отсутстве критических данных - поэтому и удаляем.

### Про ортогонализацию параметров частоты посещений (дополнительная глава)

Несмотря на полученное выше подтверждение принципиальной двумерности пространства частот посещений, в котором мы предсказываем целевую переменную, коррелятор текущих базовых векторов 95% !!  То есть наш базис неудачно выбран и может приводить к проблемам при тренировке модели в нем - базис квазивырожден.  Я точно не знаю, как на самом деле реализован алгоритм моделей - может быть там заложена ортогонализация базиса параметров - поэтому в данной дополнительной главе посмотрим, что дает попытка такой ортогонализации простыми способами в ручную.

Причина вырожденности базиса очевидна - оба вектора содержат общий уровень величины, от которого отсчитывается изменение в последний месяц.  У одного клиента это будет, например: 2 посещения в месяц обычно и 1 посещение в месяц перед уходом, а у другого - 4 обычно и 2 перед уходом, поэтому по значению 2 посещения в последний месяц нельзя сказать, кто из них чернер, а вот если мерять относительное изменение, то у обоих будет -50% в последний месяц, и ясно, что оба - кандидаты чернеры.

Вариантом менее вырожденного базиса были бы пары - вместо (средняя частота, частота в последний месяц):
1. `abs` (средняя частота, абсолютное изменение частоты в последний месяц) = (mo_freq - tot_freq)
2. `rel` (средняя частота, относительное изменение частоты в последний месяц) = лучше будет взять (mo_freq - tot_freq) / (mo_freq + tot_freq)

Посмотрим, какие кореляционные матрицы мы увидим в этом случае:

#### Ортогонализация частотных параметров - варианты `abs` и  `rel`

```python
gym_abs = gym_df.copy(deep=True) # будем экспериментировать на полных копиях датасета
gym_abs['avg_class_frequency_current_month'] = gym_abs['avg_class_frequency_current_month']-gym_abs['avg_class_frequency_total']
gym_abs = gym_abs.rename(columns={'avg_class_frequency_current_month':'delta_abs'})

gym_rel = gym_df.copy(deep=True) # будем экспериментировать на полных копиях датасета
gym_rel['avg_class_frequency_current_month'] = (gym_rel['avg_class_frequency_current_month']-
                                                gym_rel['avg_class_frequency_total']) / (
                                                gym_rel['avg_class_frequency_current_month']+
                                                gym_rel['avg_class_frequency_total'])
gym_rel = gym_rel.rename(columns={'avg_class_frequency_current_month':'delta_rel'})



plt.figure(figsize=(20, 7.5))
myax1 = plt.subplot(1,2,1)
myax2 = plt.subplot(1,2,2)

sns.heatmap(abs(round(gym_abs.rename(columns=gym_col_dict).corr(),2)), 
            annot=True, ax=myax1, linewidths=0.1)
myax1.set_title('матрица корреляции признаков для варианта abs')    

sns.heatmap(abs(round(gym_rel.rename(columns=gym_col_dict).corr(),2)), 
            annot=True, ax=myax2, linewidths=0.1)
myax2.set_title('матрица корреляции признаков для варианта rel')   

plt.show()
```

#### Результаты ортогонализации частотных параметров:
1. В обоих вариантах удалось полностью избавится от взаимозависимости частотных параметров.
2. При этом существенно выросла и корреляция параметра последнего месяца с целевой переменной оттока.
3. Результаты лучше у варианта abs:
 * меньше корреляция со средней частотой посещений
 * больше корреляция с оттоком

# Модель прогнозирования оттока клиентов

## Выбор модели
### Тип модели
В силу того, что наш целевой признак — факт оттока клиента в следующем месяце, то подходяшим типом модели для нашей задачи является бинарная классификация.

### Алгоритм модели
Нашему типу модели соответствуют следующие алгоритмы с библиотечной реализацией в scikit-learn:
* Логистическая регрессия;
* Дерево принятия решений;
* Случайный лес;
* Градиентный бустинг.

Где последние три образуют группу по увеличению сложности алгоритма для повышения точности и скорости.  Будем исследовать только один из них - случайный лес - как более точный с более надежной сходимостью (так как размерность датасета у нас невелика и можно пожертвовать скоростью алгоритма ради надежности)

В итоге будем рассматривать два алгоритма - представителя регрессии бинарного выбора и представителя дерева решений:
1. Логистическая регрессия - конструктор модели `sklearn.linear_model.LogisticRegression()`;
2. Случайный лес - конструктор модели `sklearn.ensemble.RandomForestClassifier()`.

## Разделение датасета на обучающую и валидационную выборки

```python
X_df = gym_df.drop(columns='churn') # признаки по которым делаем предсказание
y_sr = gym_df['churn'] # целевой параметр

# разделяем данные на обучающую и валидационную выборку в соотношении 80:20
X_train, X_test, y_train, y_test = train_test_split(X_df, y_sr, test_size=0.2, random_state=0)
```

## Нормализация обучающей выборки и нормирование валидационной выборки

```python
# создадим объект класса StandardScaler 
scaler = StandardScaler()

# обучим scaler и одновременно нормализуем матрицу признаков для обучающей выборки
X_train_st = scaler.fit_transform(X_train)

# обученным scaler-ом стандартизируем матрицу признаков для тестовой выборки
X_test_st = scaler.transform(X_test)
```

## Обучение моделей на обучающей выборке - два алгоритма

```python
# Алгоритм логистической регрессии
model_R = LogisticRegression(random_state=0)
model_R.fit(X_train_st, y_train)

y_pred_R = model_R.predict(X_test_st) # предсказание в виде выбора бинарного значения
y_prob_R = model_R.predict_proba(X_test_st)[1] # предсказание в виде вероятностей бинарного значения 1 для метрик типа ROC-AUC
```

```python
# Алгоритм случайного леса
model_F = RandomForestClassifier(random_state=0)
model_F.fit(X_train_st, y_train)

y_pred_F = model_F.predict(X_test_st) # предсказание в виде выбора бинарного значения
y_prob_F = model_F.predict_proba(X_test_st)[1] # предсказание в виде вероятностей бинарного значения 1 для метрик типа ROC-AUC
```

## Сравнение точности моделей на базе метрик accuracy, precision и recall на валидационной выборке
Метрики бинарной классификации, выбранные для сравнения, базируются на матрице ошибок, для которой реализована библиотечная функция `sklearn.metrics.confusion_matrix()`.  Реализуем функцию, которая будет возвращать кортедж искомых метрик:

```python
def est(test, pred):
    TN, FP, FN, TP = confusion_matrix(test, pred).ravel()
    accuracy = ( TP + TN ) / ( TP + TN + FP + FN)
    precision = TP / ( TP + FP )
    recall = TP / ( TP + FN )
    return accuracy, precision, recall
```

Оформим результаты сравнения в виде таблицы

```python
show_df = pd.DataFrame( [ list(est(y_test, y_pred_R)), list(est(y_test, y_pred_F)) ] )
show_df.columns = ['accuracy','precision','recall']
show_df.index = ['логистическая регрессия','случайный лес']
print('\n СРАВНЕНИЕ ДВУХ АЛГОРИТМОВ ПРЕДСКАЗАТЕЛЬНОЙ МОДЕЛИ ОТТОКА')
round(show_df,2)
```

## Выводы по предсказательной модели:
1. Алгоритм случайного леса заметно точнее алгоритма логистической регрессии по всем метрикам точности, поэтому мы выбираем модель по алгоритму случайного леса для предскажания оттока в следующем месяце.
2. Наша модель ошибается в персональных предсказаниях ухода конкретного клиента только в 5% случаев.
 * при этом из всех предсказанных чернеров только 5% таковыми не станут в следующем месяце,
 * при этом наша модель обнаруживает 85% реальных чернеров за месяц до их ухода.

# Кластеризация клиентов без учета статуса оттока

## Важное обоснование сужения подмножества признаков для кластеризации (дополнительная глава)
Идея кластеризации в том, чтобы провести сегментацию клиентов с целью дать бизнесу и маркетингу простой способ интерпретации рекомендаций по борьбе с оттоком. Поэтому кластеризацию записей придется делать без столбца `churn`, однако это не единственное ограничение, которое следует ввести для успешного достижения целей кластеризации - рассмотрим факты:
* алгоритм автоматической кластеризации не знает, что кластеризацию нужно делать именно для определения сегментов с разной долей оттока.
* для работы алгоритма вводится метрика на пространстве стандартизированных значений параметров - по этому, грубо говоря, расстояния по оси любого бинарного признака будут всегда оцениваться алгоритмом кластеризации даже чуть больше, чем расстояния по оси многозначного признака - а теперь взглянем, какие признаки важны для предсказания оттока: 

```python
def line_f_importance(w):
    if   w > 0.2:  return 'huge'
    elif w > 0.1:  return 'high'
    elif w > 0.05: return 'medium'
    elif w > 0.01: return 'low'
    else:          return 'neglectable'

show_df['importance_for_churn_prediiction'] = show_df['weight'].apply(line_f_importance)
show_df = show_df.reset_index()
show_df['nof_unique_values_of_parameter'] = show_df['parameter'].apply(lambda s: gym_df[s].nunique())
show_df = show_df.set_index('parameter')
round(show_df.sort_values('weight', ascending=False),3)
```

* Таким образом наш автоматический алгоритм кластеризации будет растаскивать кластеры преимущественно по бинарным признакам, которые не влияют на отток - **зачем нам такая сегментация клиентов для задачи борьбы с оттоком?!**  Ну поделит он нам клиентов на близких и дальних (или мужчин и женщин) - и что на базе такого разделения мы посоветуем в результате маркетингу? Для сегментов, не отличающихся друг от друга по среднему значению целевого признака оттока!
* Можно сразу предложить два способа повлиять на преференции кластеризации по признакам в пользу тех, что влияют на отток:
1. после нормализации добавить веса модели в метрику расстояния для алгоритма кластеризации - тогда растаскивание по параметрым важным для оттока будет восприниматься алгоритмом как более эффективное,
2. просто отбросить все параметры со значением меньше 1% (neglectable).

В первом случае кластеризация будет лучше всего сбалансирована, во-втором сильно упростится, сохранив при этом преференции по оттоку.  Оптимальным был бы комбинированный вариант:

3. сперва отбросить все neglectable, а потом использовать веса оставшихся признаков как гео-коэффициенты для метрики расстояния вдоль осей данных признаков при кластеризации.

**Выбор**:  будь мой опыт в ML побольше, чем неделя теории спринта и неделя данного проекта, я бы выбрал вариант 3.  Однако у меня есть опасения, что мой текущий уровень владения предметной областью может завести меня на самом сложном пути в ситуацию, разрешить которую мне будет не по силам, поэтому выбираем для дальнейшей работы **ВАРИАТ 2: будем использовать стандартизированную метрику с равными весами, но удалим из рассмотрения для кластеризации не только поле оттока, а и все поля neglectable из таблицы выше**:

Наверное, лучше сформулировать совет так: если бюджета целевого маркетинга не хватает на всех, то лучше сфокусировать его на тех, что уже дольше посещает фитнес-клуб, чем на новых.
 * **недисциплинированное болото** это основное поле борьбы с черном среди уже привлеченных клиентов.  Их основное отличие - самая низкая частота посещения фитнес-центра: ниже даже чем у гарантированных чернеров!  Я могу предположить, что 30% чернеров среди них - это тоже принявшие решение об уходе, но выхаживающие свой абонемент до конца (в отличие от первой группы, которые уходят, теряя деньги подписки последнего месяца.  Остальные 70% из них - наша цель.  Поскольку других выдающихся значений признаков у этого сегмента нет, речь идет просто об отсутствии мотивации чаще заниматься.  Я аналитик-отличник и не знаю чем поднимается мотивация ленивых троечников - нужен совет психолога и идеи самого маркетинга по тому, как их перетянуть в группу ниже.
 * **фанаты-качки** (ранее *дисциплинированные клиенты с обычными абонементами*) - такое название дано, поскольку эта группа является антиподом предыдущей, у них схожие значения всех параметров, за исключением мотивации заниматься - частота посещения занятий 3 раза в неделю в среднем, максимальная среди всех сегментов! 
 * **корпоративные клиенты с годовым абонементом** (ранее *дисциплинированные клиенты с годовым абонементом*) - название отредактированно, так как замечено, что хотя кластеризация и не учитывала слабо-коррелирующие с оттоком параметры, но именно у данной группы наблюдаются выдающиеся доли партнерских программ и промо-акций.  Также к особенностам этой группы относится любовь к групповым занятиям.  Я бы предположил, что все это результат хорошей работы отдела b2b продаж, включающей точное определение целевой аудитории и умелое применение промо-акций среди нее.  В целом можно посоветовать растить b2b-подход к продажам, разрабатывать и предлагать партнерские программы близрасположенным офисам крупных компаний.
